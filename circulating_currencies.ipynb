```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
url="https://en.wikipedia.org/wiki/List_of_circulating_currencies"
page=requests.get(url)
soup=BeautifulSoup(page.content,'html5lib')
soup
```

```python
tdata=soup.find_all(class_="wikitable sortable")
len(tdata)
```

```python
state_data=tdata[0]
state_data
```

```python
all_rows=state_data.find_all('tr')
len(all_rows)
```

```python
heads=[d.get_text().strip() for d in all_rows[0].find_all("th")]
heads
```

```python
data= [[h.get_text().strip() for h in d.find_all('td')] for d in all_rows[1:]]
data
```

```python
heads
```

```python
d={k:[] for k in heads}
for i in data:
    if len(i)==6:
        d['State or territory[1]'].append(i[0])
        d['Currency[1][2]'].append(i[1])
        d['Symbol[D] orAbbrev.[3]'].append(i[2])
        d['ISO code[2]'].append(i[3])
        d['Fractionalunit'].append(i[4])
        d['Numberto basic'].append(i[5])
        cont=i[0]
    
    elif len(i)==5:
        d['State or territory[1]'].append(cont)
        d['Currency[1][2]'].append(i[0])
        d['Symbol[D] orAbbrev.[3]'].append(i[1])
        d['ISO code[2]'].append(i[2])
        d['Fractionalunit'].append(i[3])
        d['Numberto basic'].append(i[4])
d   
```

```python
df=pd.DataFrame(d)
df
```

```python
df.to_csv("currency_state.csv",index=False)
```

```python
countries_data=tdata[1]
all_rows=countries_data.find_all('tr')
heads=[d.get_text().strip() for d in all_rows[0].find_all("th")]
data= [[h.get_text().strip() for h in d.find_all('td')] for d in all_rows[1:]]

d={k:[] for k in heads}
for i in data:
    if len(i)==7:
        d['Currency'].append(i[0])
        d['Symbol or\nAbbrev.'].append(i[1])
        d['ISO code'].append(i[2])
        d['Fractional\nunit'].append(i[3])
        d['Number\nto basic'].append(i[4])
        d['Number of countries/ territories'].append(i[5])
        d['Countries/ territories'].append(i[6])

d
df1=pd.DataFrame(d)
df1
```

```python
df1.to_csv("currency_countries.csv",index=False)
```
